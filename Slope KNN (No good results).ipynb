```python
import pandas as pd
import numpy as np
from mnist.loader import MNIST
import timeit

start = timeit.default_timer()

mnist_input_data_path = "C:/Users/Joseph Hwang/Desktop/MNIST/MNIST_data"
mndata = MNIST(mnist_input_data_path)
train_images, train_labels = mndata.load_training()

label_index = 1
combined_slopes = pd.DataFrame([])

for z in range(1, 101):
    with open("C:/Users/Joseph Hwang/Desktop/Sample_Output_Practice/" + str(z) + ".out") as infile:
        for i in range(36):
            next(infile)
        for line in infile:
            sourcefile = infile.read()
    sourcefile_list = sourcefile.split("\n")
    
    y = 0
    edge_weights = []
    df = pd.DataFrame([])
    time = sourcefile_list[0][131:]
    
    for x in range(len(sourcefile_list)): 
        if "Edge From Node" in sourcefile_list[x]:
            if time in sourcefile_list[x][131:] and sourcefile_list[x][90:114] != "":
                edge_weights.append(float(sourcefile_list[x][90:114]))
            elif sourcefile_list[x][90:114] != "":
                df.insert(loc=y, column=time, value=edge_weights)
                edge_weights.clear()
                time = sourcefile_list[x][131:]
                edge_weights.append(float(sourcefile_list[x][90:114]))
                y+=1
    df.insert(loc=y, column=time, value=edge_weights)
    
    df = df.T
    x = np.array(df.index).astype(float)
    slope = []
    
    for z in range(len(df.columns)):
        y = np.array(df[z])
        m, b = np.polyfit(x, y, 1)
        slope.append(m)
    slope.append("Label " + str(train_labels[label_index]))
    label_index+=1
    
    df3 = pd.DataFrame([])
    df3.insert(loc=0, column="slope", value=slope)
    df3 = df3.T
    combined_slopes = combined_slopes.append(df3)    
    
stop = timeit.default_timer()
print('Time: ', stop - start)  
```

```python
X = combined_slopes.iloc[:, :-1].values
y = combined_slopes.iloc[:, 90].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40)

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

```python
import pandas as pd
import numpy as np
from mnist.loader import MNIST

mnist_input_data_path = "C:/Users/Joseph Hwang/Desktop/MNIST/MNIST_data"
mndata = MNIST(mnist_input_data_path)
train_images, train_labels = mndata.load_training()

label_index = 0
combined_slopes = pd.DataFrame([])
combined_intercepts = pd.DataFrame([])

for z in range(0, 101):
    with open("C:/Users/Joseph Hwang/Desktop/Sample_Output_Practice/Sample_Output_1/" + str(z) + ".out") as infile:
        for i in range(36):
            next(infile)
        for line in infile:
            sourcefile = infile.read()
    sourcefile_list = sourcefile.split("\n")
    
    y = 0
    edge_weights = []
    df = pd.DataFrame([])
    time = sourcefile_list[0][131:]
    
    for x in range(len(sourcefile_list)): 
        if "Edge From Node" in sourcefile_list[x]:
            if time in sourcefile_list[x][131:] and sourcefile_list[x][90:114] != "":
                edge_weights.append(float(sourcefile_list[x][90:114]))
            elif sourcefile_list[x][90:114] != "":
                df.insert(loc=y, column=time, value=edge_weights)
                edge_weights.clear()
                time = sourcefile_list[x][131:]
                edge_weights.append(float(sourcefile_list[x][90:114]))
                y+=1
    df.insert(loc=y, column=time, value=edge_weights)
    
    df = df.T
    x = np.array(df.index).astype(float)
    slope = []
    intercept = []
    
    for z in range(len(df.columns)):
        y = np.array(df[z])
        m, b = np.polyfit(x, y, 1)
        slope.append(m)
        intercept.append(b)
    slope.append("Label " + str(train_labels[label_index]))
    intercept.append("Label " + str(train_labels[label_index]))
    label_index+=1
    
    df3 = pd.DataFrame([])
    df3.insert(loc=0, column="slope", value=slope)
    df3 = df3.T
    
    df4 = pd.DataFrame([])
    df4.insert(loc=0, column="intercept", value=intercept)
    df4 = df4.T
    
    combined_slopes = combined_slopes.append(df3)    
    combined_intercepts = combined_intercepts.append(df4)
```
